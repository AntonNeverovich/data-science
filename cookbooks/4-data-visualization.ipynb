# Визуализация данных
Важную часть инструментария аналитика данных составляют графические методы анализа даных или методы визуализации.   
Визуализация применяется прежде всего для:
* ислледования данных;
* передачи данных.     

Для графического анализа данных будем пользоваться библиотекой `mathplotlib`. В частности модулем `matplotlib.pyplot`. В самом простом случае `pypot` позволет выстраивать визуализацию шаг за шагом. ### 1. Построение линейной диаграммы
##### Задача
Требуется построить линейную диаграмму
##### Решение
Используем метод `plot()` библиотеки `matplotlib`

```python
%matplotlib inline

# Загрузить библиотеку
from matplotlib import pyplot as plt

# Создать данные для построения графика (ВВП по годам)
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# Создать линейную диаграмму: годы по оси Х, ВВП – по Y
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# Добавить название диаграммы
plt.title('Номинальный ВВП')

# Добавить подписи к оси Y
plt.ylabel('Млрд $')

# Показать линейную диаграмму
plt.show()
```

Линейный графики хорошо использовать для изображения тенденций.

#### Несколько линейных графиков с легендой

Метод `plt.plot()` можно вызывать несколько раз, чтобы показать несколько графиков на одной и тойже диаграмме. Если для каждой линии задано значение параметра `label=`, то легенда будет показана автоматически. Управлять расположением легенды можно с помощью параметра `loc=` метода `plt.legend()`.

```python
# Задать дисперсию и квадрат смещения
variance = [2**x for x in range(9)]
bias_squared = [2**x for x in range(8, -1, -1)]

# Рассчитать суммарную ошибку
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# Задачть цвет и тип графиков
plt.plot(xs, variance, 'g-', label='дисперсия')  # 'g-' – зеленая, сплошная
plt.plot(xs, bias_squared, 'r-.', label='квадрат смещения')  # 'r-.' – красная штрих-пунктирная
plt.plot(xs, total_error, 'b:', label='сумарная ошибка') # 'b:' – синяя пунктирная

# Задать расположение легенды, метки осей и название
plt.legend(loc=9)
plt.xlabel('Сложность модели')
plt.title('Компромисс между смещением и дисперсией')

# Показать график
plt.show()
```

### 2. Построение столбчатой диаграммы
##### Задача
Требуется построить столбчатую диаграмму
##### Решение
Используем метод `bar()` библиотеки `matplotlib`

```python
# Создать данные для построения столбчатой диаграммы
movies = ['Энни Холл', 'Бен-гур', 'Касабланка', 'Ганди', 'Вестсайдская\nистория']
num_oscars = [5, 11., 3, 8, 10]

# Создать столбчатую диаграмму
plt.bar(movies, num_oscars)

# Добавить название диаграммы
plt.title('Мои любимые фильмы')

# Добавить подпись к оси Y
plt.ylabel('Количество наград')

# Добавить метки для оси X, с названиями фильма в центре каждого интервала
plt.xticks([i for i, _ in enumerate(movies)], movies)

# Показать диаграмму
plt.show()
```

Столбчатые диаграммы хрошо подходят для демонстрации вариьруемости некоторой величины на определенном дискретном множестве элементов. Также столбчаная диаграмма хрошо подходит для построения гистограмм сгруппированных значений числового ряда, что позволяет наглядно исследовать характер распределения чисел в числовом ряду.

```python
# Импортировать библиотеку
from collections import Counter

# Создать числовой ряд
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10     # Дециль – десятая часть числа
histogram = Counter (decile(grade) for grade in grades)

# Создать столбчатую диаграмму
plt.bar(histogram.keys(),  
        histogram.values(),                 # Высота столбца
        9)                                  # Ширина каждого столбца равная 8

plt.axis([-6, 106, 0, 5])                   # Ось Х от -5 до 105
                                            # Ось Y от 0 до 5
plt.xticks([10 * i for i in range(11)])     # Метки по оси Х: 0, 10, 20, ... 100
plt.xlabel('Дециль')
plt.ylabel('Число студентов')
plt.title('Распределение оценок за экзамен №1')

# Показать диаграмму
plt.show()
```

Третий аргумент в методе `plt.bar()` определеяет ширину столбцов. На диаграмме выше ширина столбцов задана равной `9`.  Это позволяет оставить между столбцами некоторое расстояние, т.к. ширина интервалов равняется `10`. Центр столбцов по умолчанию находится на значении целевой переменной.    
При помощи метода `plt.axis()` было задано отображение оси $x$ в диапазоне от `-6` до `106`, чтобы столбцы $0$ и $100$ были полностью видны; а оси $y$ – в диапазоне от `0` до `5`. Метод `plt.axis()` устанавливает метки на оси $x$ — $0, 10, 20, \ldots, 100$

```python
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2013, 2014], mentions, 0.2)
plt.xticks(years)
plt.ylabel('Число упоминаний науки о данных')
plt.axis([2012.5, 2014.5, 499, 506])
plt.title('Прирост упоминаний науки о данных')
plt.show()
```

Необходимо быть аккуратней при использовании метода `plt.axis()`. Создание столбчатых диаграмм с осью $y$, которая начианется не с нуля, считается антипаттерном, поскольку легко дезориентирует. При использовании более осмысленных осей диаграмма выглядит не такой впечатсялющей.

```python
plt.bar(years, mentions, 0.2)
plt.axis([2012, 2015, 0, 600])
plt.title('Количество упоминаний науки о данных')
plt.show()
```

### 3. Построение точечной диаграммы
##### Задача
Необходимо построить точечную диаграмму
##### Решение
Используем метод `plt.scatter()` библиотеки `matploplib`

```python
# Создать набор данных
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

# Построить точечную даиграмму
plt.scatter(friends, minutes)

# Назначить метку для каждой точки
for label, friends_count, minutese_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                 xy=(friends_count, minutese_count),  # задать метку
                 xytext=(5, -5),  # и немного сместить ее
                 textcoords='offset points'
                )

# Задать название диаграммы и названия осей
plt.title('Зависимость между количеством минут и числом друзей')
plt.xlabel('Число друзей')
plt.ylabel('Время, проводимое на сайте ежедневно, мин.')

# Показать даиграмму
plt.show()
```

Точечная диаграмма или *диаграмма рассеивания* (scatterplot) лучше всего подходит для визуализации связи между двумя парными выборками данных. При размещении на точечной диаграмме сопоставимых переменных можно получить искаженную картину, если масштаб по осям будет определяться библиотекой `matplotlib` автоматически:

```python
test_1_grades = [99, 90, 85, 97, 80]  # оценки за тест № 1
test_2_grades = [100, 85, 60, 90, 70]  # оценки за тест № 2

plt.scatter(test_1_grades, test_2_grades)
plt.title('Несопоставимые оси')
plt.xlabel('Оценки за тест № 1')
plt.ylabel('Оценки за тест № 2')

plt.show()
```

Если добавим вызов метода `plt.axis('equal')`, то диаграмм более точно покажет, что большая часть вариации оценок приходятся на тест №2:

```python
plt.scatter(test_1_grades, test_2_grades)
plt.axis('equal')
plt.title('Cопоставимые оси')
plt.xlabel('Оценки за тест № 1')
plt.ylabel('Оценки за тест № 2')

plt.show()
```
